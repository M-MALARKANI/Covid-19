```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
covid_df=pd.read_csv('/content/Covid_19_cases4.csv')
```

```python
covid_df
```

```python
covid_df.head(5)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(covid_df['dateRep'], covid_df['day'], label='day', color='blue')
plt.title('COVID-19 Daily Cases')
plt.xlabel('dateRep')
plt.ylabel('day')
plt.legend()
plt.grid()
plt.xticks(rotation=20)
plt.show()
```
